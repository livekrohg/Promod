# Datahåndtering og statistikk 


## Statistikk 
- gjennomsnitt = summen av n elementer delt på n. 
  Her bruker man funksjonen "mean( )"
- standardavvik = spredning av målinger. avviket av gjennomsnittet kvadratroten av summen av (alle målingene - gjennomsnittet)ˆ2. 
  Her bruker man fubksjonen "std( )"
 

## Lesing av data
- Save as supported textfiles: lagrer som en tekstfil du kan hente (i dette tilfellet: "datahandtering.txt")
- Lagre tekstfilen i samme mappen som programmet 
- data = loadtxt('filnavn', skiprows='tall', delimiter'det du brukte', dtype='strg'

```python
from pylab import*

#innlesing av data
data = loadtxt('datahandtering.txt',skiprows=1,delimiter=',')

'''
print(data) #får mange arrayer inni en array. alle datatypene har blitt float

'''
nr = data[:,0] #får ut alle de første tallene i alle arrayene data[rad, kolonne]
hoyde = data[:,1] #får ut alle de andre tallene i alle arrayene 

plot(nr, hoyde)

#statistikk 
snitt = mean(hoyde) #regner gjennomsnitt
avvik = std(hoyde) #regner ut avvik 

print("snitt:", snitt, "standardavvik:", avvik)
```

## Regresjon 
Å finne en kurve som passer best mulig med de datapunktene vi har i et datasett

```python
from pylab import*
from sklearn.metrics import r2_score

T = array([0, 20, 40, 60, 80, 100])
sol_NH3 = array([88.5, 56.0, 34.0, 20.0, 11.0, 7.0])
sol_NaCl = array([35.7, 35.9, 36.4, 37.1, 38.0, 39.2])

'''

#regresjon av NaCl
grad = 2
reg_NaCl = polyfit(T, sol_NaCl, grad)
print(reg_NaCl) #de tallene du får ut er a, b, c ...

x = linspace(0, 150, 1000)
a = reg_NaCl[0]
b = reg_NaCl[1]
c = reg_NaCl[2]
y = a*x**2 + b*x + c

scatter(T, sol_NaCl) #plotter uten å trekke linjer mellom puktene 
plot(x,y)

'''

#regresjon av NH3

grad = 2
reg_NH3 = polyfit(T, sol_NH3, grad)

x = linspace(0, 150, 1000)
a = reg_NH3[0]
b = reg_NH3[1]
c = reg_NH3[2]
y = a*x**2 + b*x + c

scatter(T, sol_NH3)
plot(x, y)
grid()

#se om regresjonen er god 

y_R2 = a*T**2 + b*T + c
R2 = r2_score(sol_NH3, y_R2)
print(R2) #denne verdien er mellom 0 og 1, hvor 1 er perfekt. Da er modellen bra tilpasset til det datasettet du har 
```

# Numerisk derivasjon
- momentan vekstfart 
- numerisk derivasjon = fjerner grenseverdien 

```python
from pylab import*

def f(x):
    return x**2

def fder(f, x, dx = 1E-8):
    df = (f(x + dx)-f(x))/dx
    return df

print(fder(f,1))
```

- vær forsiktig med hvilke verdier du bruker. Velger du en alt for liten dx får du en feil fordi datamaskinen bruker totallssystemet. 
- Den optimale verdien for dx er 10 ^-8
- Dette er fordi man har med flytall å gjøre

## Diskret derivasjon
- derivere diskret data 

```python
from pylab import*

t = [5, 10, 15, 30, 50, 60]
s=[115, 190, 390, 600, 1000, 1200]
v = []
v.append(0)

for i in range (len(s)-1):
    sder = (s[i+1]-s[i])/(t[i+1] - t[i])
    v.append(sder)

plot(t, s)
```

## Trappeekspriment

```python
from pylab import*

data = loadtxt('trappeting.txt', delimiter=',')

tid = data[:, 0]
hoyde = data[:, 2]
fart = []
aks = []

plot(tid, hoyde, label = 'Høyde') #plotter høydegraften
show()

#gjør arrayene om til lister
tidl = list(tid)
hoydel = list(hoyde)

#diskret derivasjon
for i in range (len(hoydel)-1):
    sder = (hoydel[i+1]-hoydel[i])/(tid[i+1] - tid[i])
    fart.append(sder)

tidl.pop(-1) #tid har ett element mindre enn fart
plot(tidl, fart) #plotter fartsgrafen
show()

#diskret derivasjon
for i in range (len(fart)-1):
    vder = (fart[i+1]-fart[i])/(tid[i+1] - tid[i])
    aks.append(vder)

tidl.pop(-1) #tid har et element mindre enn akselerasjonen
plot(tidl, aks) #plotter akselerasjonsgrafen 
show()
```
